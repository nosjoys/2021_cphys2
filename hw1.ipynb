### Gradient Descent
$ f(x)={1 \over 2} x^4-3x^3$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-10,10,0.01)
def f(x):
    return 0.5*x**4-3*x**3
plt.ylim(-100,100)
plt.xlim(-5,7)
plt.plot(x,f(x))
```

```python
x=6
x_old=1
eps=0.01
precision=0.00000001
def fp(x):
    return 2*x**3-9*x**2
while (abs(x-x_old)>precision):
    x_old=x
    x=x_old-eps*fp(x_old)
```

```python
x,f(x)   #최저점
```

## Diabetes (regression)

```python
from sklearn.linear_model import LinearRegression
from sklearn import datasets
import matplotlib.pyplot as plt
import numpy as np
```

```python
diabetes=datasets.load_diabetes()
```

```python
diabetes.keys()
```

```python
diabetes.DESCR
```

```python
X_age=diabetes.data[:,0] # 나이
X_bmi=diabetes.data[:,2] # BMI
X_p=diabetes.data[:,3] # 혈압
Y = diabetes.target
```

 #### 나이, 진행정도 (scikit learn 사용 x)

```python
def f(a,b,x):
    return a*x+b

def mse(a,b,y):
    y_hat=f(a,b,X_age)
    return np.mean((y_hat-y)**2)
```

```python
a=500; b=150; n=0; d=1;
while (n<10000):
    if mse(a+d,b,Y)>mse(a-d,b,Y):
        a=a-d
    else:
        a=a+d
    if mse(a,b+d,Y)>mse(a,b-d,Y):
        b=b-d
    else:
        b=b+d
    n=n+1
```

```python
plt.plot(X_age,Y,'o')
plt.plot(X_age,f(a,b,X_age))
a,b
```

 #### 나이, 진행정도 (scikit learn 사용 o)

```python
f=LinearRegression()
X=[[i] for i in X_age]
f.fit(X,Y)
```

```python
f.coef_,f.intercept_
```

```python
plt.plot(X_age,Y,'o')
plt.plot(X,f.predict(X))
```
